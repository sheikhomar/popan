```python
def plot_random(images, image_shape=(30, 40)):
    num_rows = 2
    num_columns = 7
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_columns, sharex=True, sharey=True, figsize=(9, 4))
    plt.subplots_adjust(hspace=0.1, wspace=0.1) 
    for row in range(num_rows):
        for column in range(num_columns):
            img = images[column + (row*num_columns)]
            img = img.reshape(image_shape).T
            ax = axes[row][column]
            ax.imshow(img, cmap='gist_gray')
            #ax.set_axis_off()
            ax.tick_params(axis='both', which='major', bottom=False, labelbottom=False, left=False, labelleft=False)
```

```python
plot_random(X_train)
```
