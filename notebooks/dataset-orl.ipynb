```python
def get_subjects(X, y, filter_labels, select_one=True):
    final_list = None
    for label in filter_labels:
        filtered = X[y==label][0].reshape(1, -1) if select_one else X[y==label]
        if final_list is None:
            final_list = filtered
        else:
            final_list = np.concatenate([final_list, filtered], axis=0)
    return final_list
```

```python
subjects = get_subjects(X, y, np.unique(y), select_one=True)
utils.plot_images(subjects, n_rows=4, n_columns=10);
```
